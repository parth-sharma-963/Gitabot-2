```python
# final_code_advanced.py

import json
import re
import numpy as np
from sentence_transformers import SentenceTransformer, util
from typing import List, Dict, Any

# A more comprehensive set of common English words to ignore for better keyword matching
STOPWORDS = {
    "i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours",
    "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers",
    "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves",
    "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are",
    "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does",
    "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until",
    "while", "of", "at", "by", "for", "with", "about", "against", "between", "into",
    "through", "during", "before", "after", "above", "below", "to", "from", "up", "down",
    "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here",
    "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more",
    "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so",
    "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"
}

# The name of the pre-trained model for semantic understanding
MODEL_NAME = 'all-MiniLM-L6-v2'

def load_gita_dataset(file_path: str) -> List[Dict[str, Any]]:
    """Loads the Bhagavad Gita dataset from a JSON file."""
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            return json.load(file)
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        print("Please make sure the JSON file is in the correct directory.")
        exit()
    except json.JSONDecodeError:
        print(f"Error: The file '{file_path}' is not a valid JSON file.")
        exit()

def preprocess_and_embed_dataset(dataset: List[Dict[str, Any]]):
    """
    Prepares the dataset for searching by creating a combined text field for each verse
    and generating semantic embeddings using a transformer model.
    """
    print("Pre-processing dataset and generating semantic embeddings...")
    print(f"This might take a moment as the model '{MODEL_NAME}' is loaded...")

    # Load the sentence transformer model
    model = SentenceTransformer(MODEL_NAME)

    # Create a comprehensive text representation for each verse for better semantic search
    for item in dataset:
        themes_str = ", ".join(item.get("themes", []))
        keywords_str = ", ".join(item.get("keywords", []))
        item['combined_search_text'] = (
            f"Theme: {themes_str}. Keywords: {keywords_str}. "
            f"Context: {item.get('context', '')}. "
            f"Translation: {item.get('translation', '')}"
        )

    # Generate embeddings for all verses in the dataset
    corpus_embeddings = model.encode(
        [item['combined_search_text'] for item in dataset],
        convert_to_tensor=True,
        show_progress_bar=True
    )

    print("Embeddings generated successfully. Ready for your questions.")
    return model, corpus_embeddings

def extract_keywords(text: str) -> set:
    """Extracts meaningful keywords from user input, filtering out common stopwords."""
    words = re.findall(r"\b\w+\b", text.lower())
    return {word for word in words if word not in STOPWORDS}

def search_verses_hybrid(
    dataset: List[Dict[str, Any]],
    user_query: str,
    model: SentenceTransformer,
    corpus_embeddings
) -> List[Dict[str, Any]]:
    """
    Searches for verses using a hybrid approach:
    1.  **Keyword Score:** Matches user keywords against verse keywords, themes, and translation.
    2.  **Semantic Score:** Finds verses with similar meaning to the user's query.
    The scores are combined for a final relevance ranking.
    """
    user_keywords = extract_keywords(user_query)

    # 1. Calculate Keyword Scores
    keyword_scores = []
    for item in dataset:
        score = 0
        verse_content = " ".join(item.get("keywords", []) + item.get("themes", []) + [item.get("translation", "")]).lower()
        for keyword in user_keywords:
            if keyword in verse_content:
                score += 1
        keyword_scores.append(score)

    # Normalize keyword scores to be between 0 and 1
    max_kw_score = max(keyword_scores) if any(keyword_scores) else 1
    normalized_keyword_scores = [s / max_kw_score for s in keyword_scores]

    # 2. Calculate Semantic Scores
    query_embedding = model.encode(user_query, convert_to_tensor=True)
    cosine_scores = util.cos_sim(query_embedding, corpus_embeddings)[0].cpu().numpy()

    # 3. Combine scores into a hybrid score
    # We can weigh the importance of semantic vs. keyword search. Here we use 70% semantic, 30% keyword.
    hybrid_scores = [0.7 * sem_score + 0.3 * kw_score for sem_score, kw_score in zip(cosine_scores, normalized_keyword_scores)]

    # Get the indices of the top 5 scores
    top_indices = np.argsort(hybrid_scores)[-5:][::-1]

    # Prepare the results
    relevant_verses = []
    for idx in top_indices:
        verse = dataset[idx]
        verse_info = {
            "chapter": verse["chapter"],
            "verse": verse["verse"],
            "sanskrit": verse["sanskrit"],
            "translation": verse["translation"],
            "context": verse["context"],
            "relevance_score": hybrid_scores[idx]
        }
        # Only add verses with a meaningful score
        if verse_info["relevance_score"] > 0.25: # Threshold to avoid irrelevant results
            relevant_verses.append(verse_info)

    return relevant_verses

def main():
    """Main function to load the dataset, get user input, and display relevant verses."""
    # IMPORTANT: Update this path to where your JSON file is located
    dataset_path = "E:\\bhagawatGPT\\bhagavad_gita_dataset_expanded.json"

    dataset = load_gita_dataset(dataset_path)
    model, corpus_embeddings = preprocess_and_embed_dataset(dataset)

    print("\n--- Bhagavad Gita AI Assistant ---")
    print("Ask about life's challenges, emotions, or philosophical concepts.")
    print("For example: 'I am feeling anxious about the future' or 'What is the nature of the soul?'")

    while True:
        user_input = input("\n> Enter a topic or feeling (or type 'exit' to quit): ")
        if user_input.lower() in ['exit', 'quit']:
            print("\nGoodbye! May wisdom guide your path. 🙏")
            break

        if not user_input.strip():
            print("Please enter a query.")
            continue

        matched_verses = search_verses_hybrid(dataset, user_input, model, corpus_embeddings)

        if matched_verses:
            print(f"\nTop {len(matched_verses)} most relevant verses for '{user_input}':")
            for verse in matched_verses:
                print("-" * 40)
                print(f"🔹 **Chapter {verse['chapter']}, Verse {verse['verse']}** (Relevance: {verse['relevance_score']:.2f})")
                print(f"📜 **Sanskrit:** {verse['sanskrit'].strip()}")
                print(f"📖 **Translation:** {verse['translation']}")
                print(f"📚 **Context:** {verse['context']}")
        else:
            print("\nI couldn't find a closely matching verse. Please try rephrasing your question.")

if __name__ == "__main__":
    main()
```
